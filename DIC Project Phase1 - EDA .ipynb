```python
df['Fetal State Class'] = df['Fetal State Class'].astype(int)
df['CLASS'] = df['CLASS'].astype(int)
arr = df['Fetal State Class'].to_numpy()
arr2 = df['CLASS'].to_numpy()
arr
arr2
#plt.pie(arr)
#plt.show()
```

```python
def label_function(val):
    return f'{val / 100 * len(df):.0f}\n{val:.0f}%'

N = 50
df = pd.DataFrame({'CLASS': arr2,
                   'Fetal State Class': arr})

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))

df.groupby('CLASS').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 5},
                                  colors=['tomato', 'gold', 'skyblue', 'greenyellow', 'cyan', 'olive','brown','gray','pink','green'], ax=ax1)
df.groupby('Fetal State Class').size().plot(kind='pie', autopct=label_function, textprops={'fontsize': 10},
                                 colors=['violet', 'lime', 'red'], ax=ax2)
ax1.set_ylabel('According to FHR State Class Code', size=15)
ax2.set_ylabel('According to Fetal State Class', size=15)
plt.tight_layout()
plt.show()
```
